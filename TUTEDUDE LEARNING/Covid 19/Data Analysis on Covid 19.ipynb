```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('covid19_Confirmed_dataset.csv')
dataset.head()
```

```python
dataset.shape
```

```python
#Delete the useless columns
df=dataset.drop(["Lat", "Long"], axis = 1,inplace = True)
```

```python
dataset.head()
```

```python
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
import matplotlib.pyplot as plt

# Assuming 'corona_dataset_aggregated' is a DataFrame
# Check the first few rows to see the data structure
print(corona_dataset_aggregated.head())
print(corona_dataset_aggregated.columns)

# Drop the non-date column if necessary (e.g., 'Province/State')
corona_dataset_cleaned = corona_dataset_aggregated.drop(columns=['Province/State'], errors='ignore')

# Check data types to ensure the columns are numerical
print(corona_dataset_cleaned.dtypes)

# Plot the data for specific countries, ensuring we have the correct columns
corona_dataset_cleaned.loc['China'].plot(label='China')
corona_dataset_cleaned.loc['India'].plot(label='India')
corona_dataset_cleaned.loc['Spain'].plot(label='Spain')
corona_dataset_cleaned.loc['Italy'].plot(label='Italy')

plt.legend()
plt.show()
```

```python
corona_dataset_cleaned.loc["China"][:3].plot()
```

```python
countries = list(corona_dataset_cleaned.index)
max_infection_rates = []

for c in countries:
    max_infection_rates.append(corona_dataset_cleaned.loc[c].diff().max())
corona_dataset_cleaned["Max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_cleaned
```

```python
corona_data = pd.DataFrame(corona_dataset_cleaned["Max_infection_rates"])
```

```python

corona_data.plot();
```
